```python
def preprocessig(df, text_column):
    removed_punc = []
    tokens = []
    clean_tokens = []
    lemm_words = []
    clean_text = []
    df['text'] = text_column
    for line in df['text']:
        line = ''.join([i for i in line if i not in string.punctuation])
        removed_punc.append(line)
    df['removed_punc'] = removed_punc

    for line in df['removed_punc']:
        lines = nltk.word_tokenize(line)
        tokens.append(lines)
    df['tokens'] = tokens

    for line in df['tokens']:
        lines = [i for i in line if i not in stopwords.words('english')]
        clean_tokens.append(lines)
    df['clean_tokens'] = clean_tokens

    for line in df['clean_tokens']:
        lines = [WordNetLemmatizer().lemmatize(i) for i in line]
        lemm_words.append(lines)
    df['lemm_words'] = lemm_words

    for line in df['lemm_words']:
        lines = ' '.join([i for i in line if i not in string.punctuation])
        clean_text.append(lines.lower())
    df['clean_text'] = clean_text
    
    return df

df = preprocessig(df, df['text'])
df
```

# DONNER une ettequite Grammatical POS_tags

```python
import nltk
pos_tags_list = []  # initialiser une liste vide pour stocker les pos_tags de chaque liste de tokens

for tokens_list in df['lemm_words']:
    pos_tags = nltk.pos_tag(tokens_list)  # appliquer la fonction pos_tag() DONNER une ettequite Grammatical sur chaque liste de tokens
    pos_tags_list.append(pos_tags)  # ajouter les pos_tags à la liste pos_tags_list

print(pos_tags_list)  # afficher la liste de tous les pos_tags
```

# appliquer la fonction ne_chunk()

```python
import nltk
nltk.download('maxent_ne_chunker')  # télécharger les données nécessaires pour la reconnaissance d'entités nommées
nltk.download('words')  # télécharger les données nécessaires pour la reconnaissance d'entités nommées

# Assurez-vous d'avoir déjà stocké les résultats de nltk.pos_tag() dans une liste nommée pos_tags_list

for pos_tags in pos_tags_list:
    chunks = nltk.ne_chunk(pos_tags, binary=True)  # appliquer la fonction ne_chunk() sur chaque liste de pos_tags
    for chunk in chunks:
        if hasattr(chunk, 'label') and chunk.label() == 'NE':  # vérifier si l'élément est une entité nommée
            print(chunk)
```

# l'analyse de chunks

```python
# Créer deux listes vides pour stocker les entités et les labels associés
entities = []
labels = []
# Pour chaque ligne dans la colonne 'lemm_words' du dataframe df
for tokens in df['lemm_words']:
    # Obtenir les tags POS pour les tokens de la ligne actuelle
    pos_tags = nltk.pos_tag(tokens)
     # Appliquer l'analyse de chunks pour cette ligne en utilisant le POS tagging précédent
    chunks = nltk.ne_chunk(pos_tags, binary=False)
    # Pour chaque chunk résultant, vérifier si un label est associé
    for chunk in chunks:
         # Si un label est présent, ajouter l'entité associée et son label à leurs listes respectives
        if hasattr(chunk, 'label'):
            entities.append(' '.join(c[0] for c in chunk))
            labels.append(chunk.label())
# Créer une liste unique des paires entité/label (élimine les doublons)
entities_labels = list(set(zip(entities, labels)))
# Créer un dataframe à partir de la liste unique d'entités et de labels
entities_df = pd.DataFrame(entities_labels, columns=["Entities", "Labels"])

```

# Model : LogisticRegression

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(entities_df['Entities'], entities_df['Labels'], test_size=0.2, random_state=42)

# Vectorize the text data
vectorizer = TfidfVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Initialize and train the classification model
classifier = LogisticRegression()
classifier.fit(X_train_vectorized, y_train)

# Evaluate the model
accuracy = classifier.score(X_test_vectorized, y_test)
print("Model accuracy:", accuracy)
```

```python
import re
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer

def remove_punctuation(text):
    punctuationfree = "".join(
        [i for i in text if i not in string.punctuation])
    return punctuationfree


def remove_numbers(text):
    return re.sub(r'\d+', '', text)


def tokenization(text):
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(text)
    return tokens


def remove_stopwords(text):
    stopword = stopwords.words('english')
    news_stopwords = [i for i in text if i not in stopword]
    return news_stopwords


# lemmitization using WordNetLemmatizer
def lemmatization(text):
    lemmatizer = WordNetLemmatizer()
    news_lemmatization = [lemmatizer.lemmatize(i) for i in text]
    return news_lemmatization

#  Define the predict_fake_news function


def predict(text):
    text = text.lower()
    text = remove_punctuation(text)
    text = tokenization(text)
    text = remove_stopwords(text)
    text = lemmatization(text)
    text = ' '.join(text)
    return text
```
